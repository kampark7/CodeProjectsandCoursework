```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
plt.rc('figure', figsize=(10, 6))
np.set_printoptions(precision=4, suppress=True)

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

```python
data = pd.read_csv("accidents_2017.csv") 

data.head(5)
```

```python
mean = data['Victims'].mean(skipna = True) 
print("Average victim per accident: " + str(mean))
```

```python
df6 = data.groupby(['Part of the day'])['Victims'].mean()
df6
```

```python
df6.plot.barh()
```

```python
df2 = data.groupby(['District Name'])['Victims'].mean()
df2
df2.plot.barh(x='District Name', y='Victims')
```

```python
df3 = data.groupby(['Neighborhood Name'])['Victims'].mean()
df3
df3.plot.barh()
```

```python
plt.scatter(s = np.repeat(0.05, len(data)), x=data['Latitude'], y=data['Longitude'])
```

```python
df4 = data.groupby(['Hour'])['Victims'].mean()
df4
df4.plot.bar()
```
