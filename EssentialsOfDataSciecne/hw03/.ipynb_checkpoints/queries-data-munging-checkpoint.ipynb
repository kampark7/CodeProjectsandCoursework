```python
words = pd.read_sql_query("SELECT * FROM hand", con)
words
```

```python
#words.query('word == "rot"').mean()
rot = words[words.word == 'rot']
#column r value merge episode id with this data set and the tiles data set to create a smaller data set
tiles.merge(rot, left_on = 'episode_id', right_on = 'episode_id')
```

```python
links = words[words.word == 'links']
merge_links = tiles.merge(links, left_on = 'episode_id', right_on = 'episode_id')
merge_links
```

```python
mean_links = merge_links['pos_x'].mean()
print("The average of positon x for the word links is " + str(mean_links))
```

```python
not_links = words[words.word != 'links']
merge_values = tiles.merge(not_links, left_on = 'episode_id', right_on = 'episode_id')
merge_values
```

```python
mean_notlinks = merge_values['pos_x'].mean()
mean_notlinks
print("The average postion x of values that are not the word link is: " + str(mean_notlinks))
```

The average for a value that is not the word links is a bit lower than the average for the word links
