```python
data.dropna(inplace=True)
types = set(data.Team)
types
```

```python
for t in types:
    senior = data.query("Senior_Management == True and Team == '{}'".format(t))
    non_senior = data.query("Senior_Management == False and Team == '{}'".format(t))
    print("Team, Num senior: {}, num non_senior {}".format(t,len(senior), len(non_senior)))
```
