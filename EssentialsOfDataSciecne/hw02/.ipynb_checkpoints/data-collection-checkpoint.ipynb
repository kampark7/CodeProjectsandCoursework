```python
import json
import csv
from pandas.io.json import json_normalize
import numpy as np
import pandas as pd
```

```python
from lxml import etree, objectify

path = 'nursexml.xml'
parsed = objectify.parse(open(path))
root = parsed.getroot()

data = []

skip_fields = ['PARENT_SEQ', 'INDICATOR_SEQ',
               'DESIRED_CHANGE', 'DECIMAL_PLACES']

for elt in root.row.row:
    el_data = {}
    for child in elt.getchildren():
        if child.tag in skip_fields:
            continue
        el_data[child.tag] = child.text
    data.append(el_data)
data


xml_data = pd.DataFrame(data)
xml_data
```

```python
data = json.load(open('nursejson.json'))
df1 = pd.DataFrame(data["data"])
df = df1.drop(df1.iloc[:, 0:8], axis = 1)
df
```

```python
# merged_inner = pd.merge(left=xml_data,right=df, left_on='state_or_nation', right_on=df.columns[0])
# merged_inner.head()
joindf = xml_data.join(df.set_index(df.columns[0]), on='state_or_nation', how='inner')
joindf.head()
```

```python
joindf.to_csv('joined_data.csv')
```
