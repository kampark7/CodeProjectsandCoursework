```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn import linear_model
```

```python
data = pd.read_csv('Weather.csv')
```

```python
data['DidItRain'] = data['RainTotal'].apply(lambda x: 1 if x > 0 else 0)
```

```python
data
```

```python
feature_cols = ['TempAvg','DewAvg','HumidityAvg','WindAvg','PressureAvg'] 
#decided to do averages since someone would probably know those features better than the max/min for the day
X = data[feature_cols] # Features
y = data.DidItRain # Target variable
```

```python
X_train,X_test,y_train,y_test=train_test_split(X, y, test_size=0.25, random_state=25)
```

```python
# instantiate the model (using the default parameters)
logreg = LogisticRegression(max_iter=5000)

# fit the model with data
logreg.fit(X_train,y_train)


y_pred = logreg.predict(X_test)
```

```python
X_test
```

```python
print("Accuracy:",accuracy_score(y_test, y_pred))
```

```python
#Prediction
#['Date','TempAvg','DewAvg','HumidityAvg','WindAvg','PressureAvg']
new_date = [[51.9, 40, 75.2, 2.5, 26.9]]
#Checked an actual date that was in the test set that was true for rain but set it to 2020 and changed a few other things
prediction = logreg.predict(new_date)
prediction[0]
```
