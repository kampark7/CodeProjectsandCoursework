```python
import pandas as pd
import numpy as np
import sys
from fastai import*
from fastai.vision import*
from fastai.vision.all import*
```

```python
path = untar_data(URLs.PETS); path
```

```python
path.ls()
```

```python
files = get_image_files(path/"images")
len(files)
```

```python
np.random.seed(2)
pat = r'^(.*)_\d+.jpg'
batch_size = 64
```

```python
image_batch = ImageDataLoaders.from_name_re(path, files, pat,
                                          item_tfms = Resize(460),
                                          batch_size = batch_size,
                                          num_workers = 0)
```

```python
image_batch.show_batch()
```

```python
learn = cnn_learner(image_batch, resnet34, loss_func = CrossEntropyLossFlat(), metrics = error_rate)
```

```python
learn.lr_find()
```
